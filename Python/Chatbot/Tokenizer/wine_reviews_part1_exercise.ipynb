## 1.

* `description` 컬럼 정보를 조회 하여 `desc` 변수에 할당.

```python
desc = dfReviews['description']
print(desc)
```

## 2.

* `description` 컬럼에서 첫번째 값을 조회 하여 `first_description` 변수에 할당하시오

```python
first_description = dfReviews['description'].iloc[0]
print(first_description)
```

## 3. 

* `dfReviews` 에서 첫번째 row 를 선택하여 `first_row` 변수에 할당하시오.
* `first_row` 의 데이터 타입을 출력 하시오

```python
first_row = dfReviews.iloc[0]
print(type(first_row))
print(first_row)
```

## 4.
* `description` 컬럼에서 처음 10개의 데이터를 선택하여 `first_descriptions` 변수에 할당하시오.

```python
first_descriptions = dfReviews['description'].iloc[:10]
print(first_descriptions)
```

## 5.

* 라벨 인덱스의 값이 `1`, `2`, `3`, `5`, `8` 인 데이터를 선택하여 `sample_reviews` 변수에 할당 하시오

```python
sample_reviews = dfReviews.loc[[1,2,3,5,8]]
print(sample_reviews)
```

## 6.

* 라벨 인덱스의 값이`0`, `1`, `10`, `100` 인 데이터의 `country`, `province`, `region_1`, `region_2` 컬럼 정보 선택하여 `df` 변수에 할당 하시오

```python
df = dfReviews.loc[[0,1,10,100],['country','province','region_1','region_2']]
print(df)
```

## 7.

* `country`, `variety` 컬럼 정보들의 처음 100건의 데이터를 선택하여 `df` 변수에 할당 하시오. 

```python
df = dfReviews.loc[:99,['country','variety']]
print(df)
```

## 8.

* `country` 컬럼의 값이 `Italy` 인 리뷰 정보를 선택하여 `italian_wines` 변수에 할당하시오

```python
italian_wines = dfReviews.loc[dfReviews['country'] == 'Italy']
print(italian_wines)
```

## 9.

* Australia 와 New Zealand 에서 만들어진 와인중에 `point` 가 95점 이상인 데이터를 선택하여 `top_oceania_wines` 변수에 할당하시오

```python
top_oceania_wines = dfReviews.loc[
                                  (
                                      (dfReviews['country'] == 'Australia') | 
                                   (dfReviews['country'] == 'New Zealand')
                                   ) & 
                                  (dfReviews['points'] >= 95)
                                  ]
print(top_oceania_wines)
```
