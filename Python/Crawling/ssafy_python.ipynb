```
apps_of_ages = {'4+':4433, '9+':987, '12+':1155, '17+':622}
total_apps = 7197
ratios = {}
percentages = {}

for apps_of_age in apps_of_ages:
  ratios[apps_of_age] = apps_of_ages[apps_of_age]/total_apps
  percentages[apps_of_age] = apps_of_ages[apps_of_age]/total_apps*100

print(ratios)
print(percentages)
```

```
print(ratios['17+'])

sum = 0
for apps_of_age in percentages:
  if int(apps_of_age.replace('+','')) <= 15:
    sum += percentages[apps_of_age]

print(sum)
```

```
def avgFunc(*scores):
  sum=0
  for score in scores:
    sum += score
  avg_score = sum/len(scores)
  return avg_score

print(avgFunc(84,69,75))
print(avgFunc(87,90,89,95))
```

```
def grade(*scores):
  sum=0
  for score in scores:
    sum += score
  avg_score = sum/len(scores)

  if avg_score >= 90:
    return "A"
  elif avg_score >=80:
    return "B"
  elif avg_score >=70:
    return "C"
  else:
    return "F"

print(grade(84,69,75))
print(grade(87,90,89,95))
```

```
def grade2(avg_score):
  if avg_score >= 90:
    return "A"
  elif avg_score >=80:
    return "B"
  elif avg_score >=70:
    return "C"
  else:
    return "F"

print(grade2(avgFunc(84,69,75)))
print(grade2(avgFunc(87,90,89,95)))
```
