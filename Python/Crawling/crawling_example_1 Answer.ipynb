### 크롤링을 위한 라이브러리 import

```
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

### Q1. 미국 Customer Reports 사이트에서 아래 url로 제품 정보(html) 조회
* 'https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
* UserAgent 라이브러리를 이용해 chrome 에서 접속한 것처럼 header 정보 생성

```
user_agent = UserAgent()
main_url = 'https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
page = requests.get(main_url,headers={'user-agent':user_agent.chrome})
soup = BeautifulSoup(page.content,'lxml')
```

### Q2. 제품의 이름 정보와 링크 정보 조회 하여 딕셔너리 객체(products)에 할당

```
products = {}   # product {name - key : product link – value}

product_names = [div.a.string for div in soup.find_all('div',class_='crux-body-copy')]
product_links = [div.a['href'] for div in soup.find_all('div',class_='crux-body-copy')]

products = {div.a.string:div.a['href'] for div in soup.find_all('div',class_= 'crux-body-copy')}
print(product_links)

```

### Q3. products 객체를 loop 돌면서 제품명 과 link 정보 출력 

```
for key,value in products.items():
  print(key , '    -->',value)
```
