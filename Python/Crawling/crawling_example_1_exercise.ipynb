### 크롤링을 위한 라이브러리 import

```
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

### Q1. 미국 Customer Reports 사이트에서 아래 url로 제품 정보(html) 조회
* 'https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
* UserAgent 라이브러리를 이용해 chrome 에서 접속한 것처럼 header 정보 생성

```
ua = UserAgent()
header = {'user-agent':ua.chrome}
page = requests.get('https://www.consumerreports.org/cro/a-to-z-index/products/index.htm',headers=header)
```

### Q2. 제품의 이름 정보와 링크 정보 조회 하여 딕셔너리 객체(products)에 할당

```
soup = BeautifulSoup(page.content,'lxml') # html 문서가 담겨있는 content를 가져옴
products = {}
nameInfos = soup.find_all(class_="products-a-z__results__item")
for nameInfo in nameInfos:
  name = nameInfo.string.replace('\n', '')
  pHref = nameInfo['href'].replace('\n','')
  products[name] = pHref

# products = [div.a.string:div.a['href'] for div in soup.find_all('div', class_='crux-body-copy')]
```

### Q3. products 객체를 loop 돌면서 제품명 과 link 정보 출력 

```
for name,href in products.items():
  print(name, href)

#for key in products:
#  print(key, '   -->', products[key])
```
